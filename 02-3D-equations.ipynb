```python
import k3d
import numpy as np
```

https://k3d-jupyter.readthedocs.io/

```python
Nx, Ny  = 34, 33
xmin, xmax = -10, 10
ymin, ymax = -10, 10

x = np.linspace(xmin, xmax, Nx)
y = np.linspace(ymin, ymax, Ny)
x, y = np.meshgrid(x, y)

plot = k3d.plot()

f = x*0
plt_surface = k3d.surface(f.astype(np.float32), bounds=[xmin,xmax,ymin,ymax], color=0xF39C12)
plot += plt_surface

plt_text = k3d.text2d('2x-y=0', position=[0.0, 0.0], color=0xF39C12, size=1.5)
plot += plt_text

f = -x+2*y+1
plt_surface = k3d.surface(f.astype(np.float32), bounds=[xmin,xmax,ymin,ymax], color=0x2E86C1)
plot += plt_surface

plt_text = k3d.text2d('-x+2y-z=-1', position=[0.0, 0.1], color=0x2E86C1, size=1.5)
plot += plt_text

f = (4+3*y)/2
plt_surface = k3d.surface(f.astype(np.float32), bounds=[xmin,xmax,ymin,ymax], color=0xEC7063)
plot += plt_surface

plt_text = k3d.text2d('-3y+2z=4', position=[0.0, 0.2], color=0xEC7063, size=1.5)
plot += plt_text

plot.display()
```

```python
plot.fetch_snapshot()
with open('../planes.html','w') as fp:
    fp.write(plot.snapshot)
```

```python
Nx, Ny  = 34, 33
xmin, xmax = -10, 10
ymin, ymax = -10, 10

plot = k3d.plot()

x = np.linspace(xmin, xmax, Nx)
y = np.linspace(ymin, ymax, Ny)
x, y = np.meshgrid(x, y)

origin = np.array([0, 0, 0])
a = np.array([2, -1, 0])
b = np.array([-1, 2, -3])
c = np.array([0,-1, 4])

arrow = np.array([origin, a])
line = k3d.line(vertices=arrow, color=0xF39C12)
plot += line

plt_text = k3d.text2d('[2,-1,0]^{T}', position=[0.0, 0.2], color=0xF39C12, size=2)
plot += plt_text

arrow = np.array([origin, b])
line = k3d.line(vertices=arrow, color=0x2E86C1)
plot += line

plt_text = k3d.text2d('[-1,2,-3]^{T}', position=[0.0, 0.4], color=0x2E86C1, size=1.5)
plot += plt_text

arrow = np.array([origin, c])
line = k3d.line(vertices=arrow, color=0xEC7063)
plot += line

plt_text = k3d.text2d('[0,-1,4]^{T}', position=[0.0, 0.6], color=0xEC7063, size=1.5)
plot += plt_text

arrow = np.array([a, a+b])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([a, a+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([a+b, a+b+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([b, a+b])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([b, b+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([c, b+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([c, a+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([a+c, a+b+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([b+c, a+b+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

arrow = np.array([a+b, a+b+c])
line = k3d.line(vertices=arrow, color=0xAEB6BF)
plot += line

point = k3d.points([[1, 0, 1]], point_size=0.2, color=0x000000)
plot += point

plot.display()
```

```python
plot.fetch_snapshot()
with open('../vectors.html','w') as fp:
    fp.write(plot.snapshot)
```
